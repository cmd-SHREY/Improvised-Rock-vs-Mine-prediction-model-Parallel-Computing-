**IMPROVING THE RUNTIME OF THE PROGRAM**

**Importing the dependencies:-**

Importing the **Numba** Library and **jit** function

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

from numba import jit
```

Data Collection and Data Processing

**Calculating the initial time of execution using %time**

```python
%time
#loading the dataset to a pandas Dataframe
sonar_data = pd .read_csv('/content/Copy of sonar data.csv', header = None)
```

**Improvising** the **sonar_data_mean function** for better runtime
Clearly the **wall time** of execution is reduced.

```python
%time

@jit
def sonar_data_mean():
    return sonar_data.groupby(60).mean()
```

```python
# Seperating Data & Labels
X = sonar_data.drop(columns=60, axis=1)
Y = sonar_data[60]
```

```python
print(X)
print(Y)
```

```python
%time
# Spliting into Training & Test data:-

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

**Spliting** the **linear Execution** to **parallel execution** is not always profitable. Hence for smaller functions with **less overhead** Linear execution is **profitable**.

```python
%time
from numba import jit
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, stratify=Y, random_state=1)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
print(X_train)
print(Y_train)
```

MODEL TRAINING -> LOGISTIC REGRESSION

```python
%time
model = LogisticRegression()
```

```python
# training the logistic regression model with training data

model.fit(X_train, Y_train)
```

Model Evaluation

Here It can be clearly seen that the total runtime has been **3 micro seconds**

```python
%time
#accuracy on the training data:-
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print("Accuracy on training data: ", training_data_accuracy)
```

```python
%time
#accuracy on the test data:-
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
%time
from numba import jit

@jit
def accuracy_test():
    #accuracy on the test data:-
    X_test_prediction = model.predict(X_test)
    test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
    return test_data_accuracy

test_data_accuracy = accuracy_test()
```

```python
print("Accuracy on test data: ", test_data_accuracy)
```
